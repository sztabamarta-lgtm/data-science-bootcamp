```python
srednia_cena_2 = merged_columns_inner.groupby('Nazwa_Produktu')[['Srednia_Cena_Wyliczona', 'Cena_Katalogowa']].mean().reset_index()
srednia_cena_2
```

```python
matched_vaulues_in_merged_columns['Wartosc_wyliczona'] = matched_vaulues_in_merged_columns.Wartosc_Sprzedazy/matched_vaulues_in_merged_columns.Ilosc_Sprzedana
```

```python
matched_vaulues_in_merged_columns
```

```python
srednia_cena = matched_vaulues_in_merged_columns.groupby('Nazwa_Produktu')[['Wartosc_wyliczona', 'Cena_Katalogowa']].mean().reset_index()
srednia_cena
```

```python
srednia_cena['Roznica_procentowa']=(srednia_cena.Wartosc_wyliczona-srednia_cena.Cena_Katalogowa)/srednia_cena.Cena_Katalogowa *100
srednia_cena
```

```python
srednia_cena_2['Roznica_procentowa']=(srednia_cena_2.Srednia_Cena_Wyliczona-srednia_cena_2.Cena_Katalogowa)/srednia_cena_2.Cena_Katalogowa *100
srednia_cena_2
```

```python
srednia_cena_sorted = srednia_cena.sort_values(by='Roznica_procentowa', ascending=False)
display(srednia_cena_sorted)
```

```python
srednia_cena_sorted['Roznica_procentowa_abs'] = srednia_cena_sorted['Roznica_procentowa'].abs()
srednia_cena_sorted_abs = srednia_cena_sorted.sort_values(by='Roznica_procentowa_abs', ascending=False)
display(srednia_cena_sorted_abs)
```
