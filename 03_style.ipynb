```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-3,3,0.1)
x_sin=np.sin(x)

plt.plot(x,x_sin)
plt.show()
```

```python
plt.style.use('ggplot')
plt.plot(x,x_sin)
plt.show()
```

```python
plt.style.available
```

```python
plt.plot(x,x_sin)
plt.style.use('dark_background')

plt.show()
```

```python
plt.plot(x,x_sin)
plt.style.use('classic')

plt.show()
```
